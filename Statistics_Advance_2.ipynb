###THEORY


1- What is hypothesis testing in statistics?

- Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves stating a hypothesis, collecting data, and determining whether the data support or contradict the hypothesis.

2- What is the null hypothesis, and how does it differ from the alternative hypothesis?

- Null Hypothesis (H‚ÇÄ): The default assumption that there is no effect or no difference.

Alternative Hypothesis (H‚ÇÅ or Ha): The statement that contradicts the null; it represents what you want to prove.

Example:

H‚ÇÄ: Mean = 50

H‚ÇÅ: Mean ‚â† 50

3- What is the significance level in hypothesis testing, and why is it important?

- The significance level (Œ±) is the probability of rejecting the null hypothesis when it is actually true (Type I error).

Common values: 0.05 (5%), 0.01 (1%)

It's important because it sets the threshold for deciding if results are statistically significant.



4- What does a P-value represent in hypothesis testing?

- The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true.

5- How do you interpret the P-value in hypothesis testing?

- If P-value ‚â§ Œ± ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ)

If P-value > Œ± ‚Üí Fail to reject H‚ÇÄ (not enough evidence)

Example:
If Œ± = 0.05 and P-value = 0.03 ‚Üí Reject H‚ÇÄ

6- What are Type 1 and Type 2 errors in hypothesis testing?

- Type I Error (Œ±): Rejecting H‚ÇÄ when it's true

Type II Error (Œ≤): Failing to reject H‚ÇÄ when it's false

Error Type	Meaning	Consequence
Type I	False positive	Unnecessary action
Type II	False negative	Missed opportunity

7- What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

- One-tailed test: Tests for deviation in one direction (e.g., greater than or less than)

Two-tailed test: Tests for deviation in both directions (e.g., not equal to)

8- What is the Z-test, and when is it used in hypothesis testing?

- The Z-test is used to test hypotheses about population means when:

The sample size is large (n > 30)

Population standard deviation is known

19- What is the F-distribution, and when is it used in hypothesis testing?

- The F-distribution is used to compare two variances or in ANOVA. It is right-skewed and depends on two degrees of freedom.

20- What is an ANOVA test, and what are its assumptions?

- ANOVA (Analysis of Variance) tests if three or more group means are significantly different.

Assumptions:

Independence of observations

Normal distribution

Equal variances (homogeneity of variance)

21- What are the different types of ANOVA tests?

- One-way ANOVA: One independent variable

Two-way ANOVA: Two independent variables

Repeated Measures ANOVA: Same subjects measured repeatedly

- An F-test compares variances between groups.

3- Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
def one_sample_z_test(sample, population_mean, population_std):
    import numpy as np
    from scipy.stats import norm

    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = [58, 60, 62, 63, 61, 64, 65]
z, p = one_sample_z_test(sample, 60, 3)
print(f"Z-score: {z:.2f}, P-value: {p:.4f}")
```

4- Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z_stat = 2.1
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_stat, color='blue', linestyle='--', label=f"Z-stat = {z_stat}")
plt.title("Two-tailed Z-test Visualization")
plt.legend()
plt.grid(True)
plt.show()
```

5- Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu0 = 50  # Null hypothesis
mu1 = 53  # Alternative hypothesis
sigma = 5
alpha = 0.05
n = 30

se = sigma / np.sqrt(n)
z_alpha = norm.ppf(1 - alpha)
x = np.linspace(40, 60, 1000)

# Distributions
h0_pdf = norm.pdf(x, mu0, se)
h1_pdf = norm.pdf(x, mu1, se)

plt.plot(x, h0_pdf, label='H0 Distribution')
plt.plot(x, h1_pdf, label='H1 Distribution')
plt.axvline(mu0 + z_alpha * se, color='red', linestyle='--', label='Critical Value')
plt.fill_between(x, h0_pdf, where=x >= mu0 + z_alpha * se, color='red', alpha=0.3, label='Type I Error')
plt.fill_between(x, h1_pdf, where=x < mu0 + z_alpha * se, color='orange', alpha=0.3, label='Type II Error')
plt.title("Type I and Type II Errors")
plt.legend()
plt.grid(True)
plt.show()
```

9- Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
from scipy.stats import norm
import numpy as np

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = np.std(data, ddof=1) / np.sqrt(len(data))
    z = norm.ppf((1 + confidence) / 2)
    return (mean - z*se, mean + z*se)

sample = np.random.normal(100, 10, 50)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

11- Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: Disease testing
p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05

posterior = bayes_theorem(p_disease, p_pos_given_disease, p_pos_given_no_disease)
print(f"Probability of disease given positive test: {posterior:.4f}")
```
